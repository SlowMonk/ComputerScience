# stack 
LIFO(last input first output)

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Stack:
    def __init__(self):
        self.head = None

    def is_empty(self):
        if not self.head:
            return True

        return False

    def push(self, data):
        new_node = Node(data)

        new_node.next = self.head
        self.head = new_node

    def pop(self):
        if self.is_empty():
            return None

        ret_data = self.head.data

        self.head = self.head.next

        return ret_data

    def peek(self):
        if self.is_empty():
            return None

        return self.head.data

if __name__ == "__main__":
    s = Stack()

    print(s.is_empty()) # True

    s.push(1)
    s.push(2)
    s.push(3)
    s.push(4)
    s.push(5)

    print("peek of data : {}".format(s.peek())) # 5

    while not s.is_empty():
        print(s.pop()) # 5, 4, 3, 2, 1
```

```python
def solution(bridge_length, weight, truck_weights):
    bridge = [0] * bridge_length
    count  = 0
    while len(truck_weights) != 0:
        #print(count, ":", bridge, truck_weights)
        count +=1
        if sum(bridge[1:]) + truck_weights[0] <= weight:
            bridge = bridge[1:] + [truck_weights.pop(0)]
        else:
            bridge = bridge[1:] + [0]
    return count + bridge_length
```

# final solution deque 사용 list보다 빠름 

```python
from collections import deque
```

```python
def solution(bridge_length, weight, truck_weights):
    bridge = deque([0]* bridge_length)
    count  = 0
    while len(truck_weights) != 0:
        #print(count, ":", bridge, truck_weights)
        count +=1
        #print(bridge)
        bridge.popleft()
        if (sum(bridge) + truck_weights[0]) <= weight:
            bridge.append(truck_weights.pop(0))
        else:
            bridge.append(0)
    return count + bridge_length
```

```python
print(solution(2,10,[7,4,5,6]))
```
